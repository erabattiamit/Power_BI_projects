* The Attendance column can't be a object data type it should be in integer data type, so we have to convert it. 

```python
#Checking attendance data
wc["Attendance"]
```

```python
splited_list = [i.split('.') for i in wc["Attendance"]] #splitting the values based on "."
```

```python
splited_list
```

```python
type(splited_list)
```

```python
type(splited_list[0][0])
```
